````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
def plot_categorical(df, col, top=10):
    plt.figure(figsize=(8,5))
    df[col].value_counts().head(top).plot(kind='bar')
    plt.title(f"Top {top} categories in {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
````

````python
def plot_numeric(df, col):
    fig, ax = plt.subplots(1,2, figsize=(12,5))
    sns.histplot(df[col], bins=30, ax=ax[0], kde=True)
    ax[0].set_title(f"Distribution of {col}")
    
    sns.boxplot(x=df[col], ax=ax[1])
    ax[1].set_title(f"Boxplot of {col}")
    plt.show()
````

````python
categorical_cols = [
    "Citizenship", "LegalType", "Gender", "MaritalStatus",
    "Bank", "ItemType", "VehicleType", "Province",
    "CoverType", "StatutoryRiskType"
]

for col in categorical_cols:
    print(f"--- {col} ---")
    plot_categorical(df, col)
````
